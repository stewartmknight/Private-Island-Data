```python
# Get total amount of school names
base_url = "https://api.data.gov/ed/collegescorecard/v1/schools?fields=school.name&page=50&per_page=100&api_key=TbhAxcMB42gg8vzB0pCBL2bklFc6bWUT7Roebmrq"

response = requests.get(base_url).json()
totSchools = response["metadata"]["total"]

totPages = int(totSchools/100)
totPages

# 100 Schools per API request, and x amount of requests needed for all school names.
y = 0

school_names = []
school_ids = []
degree_ints = []
school_sizes = []
distance_onlys = []

while y < totPages + 1:
    base_url = "https://api.data.gov/ed/collegescorecard/v1/schools?fields=school.name,id,school.carnegie_undergrad,latest.student.size,school.online_only&page={}&per_page=100&api_key=TbhAxcMB42gg8vzB0pCBL2bklFc6bWUT7Roebmrq".format(y)
    response = requests.get(base_url).json()
    pprint(response)
    y = y+1
    time.sleep(1)
    
    for thing in response["results"]:
        school_id = thing["id"]
        school_name = thing["school.name"]
        degree_int = thing["school.carnegie_undergrad"]
        school_size = thing["latest.student.size"]
        distance = thing["school.online_only"]
        
        school_ids.append(school_id)
        school_names.append(school_name)
        degree_ints.append(degree_int)
        school_sizes.append(school_size)
        distance_onlys.append(distance)
```

```python
df = pd.DataFrame({"School Name": school_names,
                   "School ID": school_ids,
                  "School Degree": degree_ints,
                  "School Size": school_sizes,
                  "Distance Only": distance_onlys})
df.to_csv("school_list.csv")
df
```
